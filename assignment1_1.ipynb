```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("Lenna.png")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
```

```python
# convert img to grayscale
img_gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap = "gray")
plt.axis("off")
```

```python
# create gaussian noise
mean = 0 
sigma = 15  
noise = np.random.normal(mean, sigma, img_gray.shape).astype(np.int16)

# add gaussian noise
noisy_image = img_gray.astype(np.int16) + noise

# limit that between 0 to 255
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)
```

```python
# apply gaussian noise with filter
kernel_size = (5, 5)  
sigma_x = 0  
filtered_image = cv.GaussianBlur(noisy_image, kernel_size, sigma_x)
```

```python
# to calculate remaining
noisy_int16 = noisy_image.astype(np.int16)
filtered_int16 = filtered_image.astype(np.int16)

residual_noise = noisy_int16 - filtered_int16

mean_absolute_residual = np.mean(np.abs(residual_noise))

print(f"{mean_absolute_residual:.2f}")
```

```python
# display image
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('original image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(noisy_image, cmap='gray')
plt.title('noisy image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(filtered_image, cmap='gray')
plt.title('filtered image')
plt.axis('off')

plt.show()

```
