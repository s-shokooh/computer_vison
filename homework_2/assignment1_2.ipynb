```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("Lenna.png", cv.IMREAD_GRAYSCALE)

# sobel
sobel_x = cv.Sobel(img, cv.CV_64F, 1, 0, ksize=3) 
sobel_y = cv.Sobel(img, cv.CV_64F, 0, 1, ksize=3) 

gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

gradient_norm = cv.normalize(gradient_magnitude, None, 0, 255, cv.NORM_MINMAX, cv.CV_8U)

sharpened = cv.add(img, gradient_norm)
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('original image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(gradient_norm, cmap='gray')
plt.title('sobel')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sharpened, cmap='gray')
plt.title('result image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
