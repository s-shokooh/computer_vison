```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read image
img = cv.imread('Lenna.png')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
# binary threshold
_, binary = cv.threshold(gray, 127, 255, cv.THRESH_BINARY)
```

```python
# add noise for better display
noise = np.random.randint(0, 100, binary.shape)
binary_noisy = binary.copy()
binary_noisy[noise < 3] = 255  # white noise
binary_noisy[noise > 97] = 0   # black noise
```

```python
# define kernel
kernel = np.ones((3, 3), np.uint8)

# apply cleaning
opening = cv.morphologyEx(binary_noisy, cv.MORPH_OPEN, kernel)
closing = cv.morphologyEx(opening, cv.MORPH_CLOSE, kernel)
```

```python
# show result
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.imshow(binary_noisy, cmap='gray')
plt.title('Original Binary (Noisy)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(closing, cmap='gray')
plt.title('Cleaned Binary')
plt.axis('off')

plt.tight_layout()
plt.show()
```
